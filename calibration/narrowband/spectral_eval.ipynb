```python
cm = plt.get_cmap('rainbow')

images1=[]
images2=[]
max1=[]
max2=[]
min1=[]
min2=[]
for i in range(27):
    name1='fore'+str(waves[i])+'_CAM1_0.tiff'
    name2='fore'+str(waves[i])+'_CAM2_0.tiff'
    image1=Image.open(name1)
    image2=Image.open(name2)
    im1=np.array(image1)[120,95:155]
    im2=np.array(image2)[125,110:170]
    images1.append(im1)
    images2.append(im2)
    max1.append(np.max(im1))
    min1.append(np.min(im1))
    max2.append(np.max(im2))
    min2.append(np.min(im2))

for i in range(0,26):
    plt.plot(np.divide(np.flip(images1[i]),np.max(max1)),c=cm(2*(waves2[i]/1e4 - .7)),label=str(waves2[i]/1e3))
    plt.title('Camera 1 Measurements Overlaid')
    plt.ylabel('Normalized .tiff response')
    plt.xlabel('pixel')
    #plt.legend()
plt.show()
for i in range(0,26):
    plt.plot(np.divide(np.flip(images2[i]),np.max(max2)),c=cm(2*(waves2[i]/1e4 - .7)),label=str(waves2[i]/1e3))
   # plt.clim(vmin=np.min(min1),vmax=np.max(max1))
    plt.title('Camera 2 Measurements Overlaid')
    plt.ylabel('Normalized .tiff response')
    plt.xlabel('pixel')
    #plt.legend()
plt.show()
```

```python
i1=[]
i2=[]
v1=[]
v2=[]
for i in range(27):
    name1='fore'+str(waves[i])+'_CAM1_0.tiff'
    name2='fore'+str(waves[i])+'_CAM2_0.tiff'
    image1=Image.open(name1)
    image2=Image.open(name2)
    im1=np.array(image1)[120,95:155]
    im2=np.array(image2)[125,110:170]
    index1, value1 = max(enumerate(np.flip(im1)), key=operator.itemgetter(1))
    index2, value2 = max(enumerate(np.flip(im2)), key=operator.itemgetter(1))
    i1.append(index1)
    v1.append(value1)
    i2.append(index2)
    v2.append(value2)
    
```

```python
plt.plot(np.divide(waves2,1e3),i1,'d--', c='r',label='camera 1')
plt.plot(np.divide(waves2,1e3),i2,'d--',c='b',label='camera2')
plt.xlabel('Center Wavelength')
plt.ylabel('Pixel Number of Center Wavelength')
plt.title('Location of Center Wavelengths')
plt.legend()
plt.show()
```

```python
plt.plot(np.divide(waves2,1e3),np.divide(v1,np.max(max1)),'d--', c='r',label='camera 1')
plt.plot(np.divide(waves2,1e3),np.divide(v2,np.max(max2)),'d--',c='b',label='camera2')
plt.xlabel('Center Wavelength')
plt.ylabel('Normalized Peak Response')
plt.title('Normalized Peak Response')
plt.legend()
plt.show()
```
